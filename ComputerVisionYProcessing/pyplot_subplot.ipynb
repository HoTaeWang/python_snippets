```python
# matplotlib.pyplot.subplot()
# subplot(num_rows, num_cols, plot_number)
#  - num_rows:  (row) 
#  - num_cols:  (column)
#  - plot_number:    ,     1   ....
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# 3  
plt.subplot(3, 1, 1)  # 3 1   subplot
plt.plot(x, y1)
plt.title('Sine Function')

plt.subplot(3, 1, 2)  # 3 1   subplot
plt.plot(x, y2)
plt.title('Cosine Function')

plt.subplot(3, 1, 3)  # 3 1   subplot
plt.plot(x, y3)
plt.title('Tangent Function')

plt.tight_layout()   #   ...
plt.show()
```
