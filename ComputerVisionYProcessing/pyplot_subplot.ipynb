```python
# matplotlib.pyplot.subplot()
# subplot(num_rows, num_cols, plot_number)
#  - num_rows: 그림의 행(row) 개수
#  - num_cols: 그림의 열(column)개수
#  - plot_number: 서브 플롯의 위치를 나타나며, 왼쪽 상단부터 오른쪽 하단으로 1부터 시작하여 순서대로 증가....
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# 3개의 서브플릇 생성
plt.subplot(3, 1, 1)  # 3행 1열 중 첫번째 subplot
plt.plot(x, y1)
plt.title('Sine Function')

plt.subplot(3, 1, 2)  # 3행 1열 중 두번째 subplot
plt.plot(x, y2)
plt.title('Cosine Function')

plt.subplot(3, 1, 3)  # 3행 1열 중 세번째 subplot
plt.plot(x, y3)
plt.title('Tangent Function')

plt.tight_layout()   # 서브플릇간 간격 조정...
plt.show()
```
