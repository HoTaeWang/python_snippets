```python
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open("jun_megami.jpg")

print("width = ", img.width)
print("height = ", img.height)

new_width = 2*img.width
new_height = img.height

# Extent Image Width in PILLOW
new_img = img.resize((new_width, new_height))
print("width = ", new_img.width)
print("height = ", new_img.height)

plt.subplot(3, 2, 1)
plt.imshow(img)
plt.title("Original")
plt.subplot(3, 2, 2)
plt.imshow(new_img)
plt.title("Doubled Width")
plt.show()
```

```python
# Extent Image Width in OpenCV
import cv2
image = cv2.imread("jun_megami.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
new_image = cv2.resize(image, None, fx=2, fy=1, interpolation = cv2.INTER_CUBIC)

plt.subplot(3, 2, 1)
plt.imshow(image)
plt.title("Original")
plt.subplot(3, 2, 2)
plt.imshow(new_image)
plt.title("Doubled Width")
plt.show()
```

```python
# Rotate Image
theta = -145.0
rows,cols,_ = image.shape

M = cv2.getRotationMatrix2D(center=(cols//2-1, rows//2-1), angle=theta, scale=1)
new_image = cv2.warpAffine(image, M, (cols,rows))
plt.figure(figsize=(4,4))
plt.imshow(new_image)
plt.show()
```
