```python
import cv2
import os
import matplotlib.pyplot as plt
from PIL import Image

junfile = "F7u8lrRbkAAejVB.jpg"
cwd = os.getcwd()
file_path = os.path.join(cwd, junfile)
#print(file_path)

def plot_figure(img):
    plt.figure(figsize=(4,4))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show()

img = cv2.imread(file_path)
plot_figure(img)

print("Image Type : ", type(img))
print("Image Size : ", img.size)
print("Image Shape : ", img.shape)
```

```python
# Make Blue Image
blue_jun = img.copy()
blue_jun[:, :, 1] = 0
blue_jun[:, :, 2] = 0
plot_figure(blue_jun)
```

```python
# Make Red Image
red_jun = img.copy()
red_jun[:, :, 0] = 0
red_jun[:, :, 1] = 0
plot_figure(red_jun)
```

```python
# Make green Image
green_jun = img.copy()
green_jun[:, :, 0] = 0
green_jun[:, :, 2] = 0
plot_figure(green_jun)
```
