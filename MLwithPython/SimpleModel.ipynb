```python
# Simple TensorFlow Deep Learning
# 그냥 만들어본 y=5x+3

import tensorflow as tf
import numpy as np
from tensorflow import keras

# create model
def create_model():
    xs = np.array([1., 5., 10., 15.], dtype=float)
    ys = np.array([8., 28., 53., 78.], dtype=float)
    
    model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
    model.compile(optimizer='sgd', loss='mean_squared_error')
    # Train the model
    model.fit(xs, ys, epochs=500)
    
    return model

model = create_model()

x_predit = np.array([20.0], dtype=float)

print(model.predict([x_predit]))
    
```
